```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
#2.1.     :
img = cv2.imread("C:\\Users\\Aleks\\Desktop\\Phyton\\Lab_2\\test_img.jpg",)
#2.2    :
OUT_IMG_WIDTH = 500
OUT_IMG_HEIGHT = 500
#2.3   ROI  :
src = np.float32([[460, 500],[870, 500],[1250, 650],[150,650]])
#2.4   ROI  :
dst = np.float32([[0,0],[500,0],[500,500],[0,500]])
#2.5    
M = cv2.getPerspectiveTransform(src, dst)
# 2.6   
wraped_img = cv2.warpPerspective(img, M, (OUT_IMG_WIDTH,OUT_IMG_HEIGHT))
#2.7   
plt.figure(figsize=(10, 10))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(wraped_img),plt.title('Output')
plt.show()
```
