```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# **1. Методами морфологічної обробки збільшити та зменшити контури зображення**

```python
img_8 = cv2.imread("/content/1_8.png", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(5, 5))
plt.imshow(img_8, cmap='gray')
plt.show();
```



---


**Створюємо з вхідного зображення бінарне зображення**

```python
bin_img_8 = img_8.copy()
threshold = 10
max_val = 255
min_val = 0
```

```python
for row in range(img_8.shape[0]):
    for col in range(img_8.shape[1]):
        if (img_8[row][col]) > threshold:
            bin_img_8[row][col] = max_val
        else:
            bin_img_8[row][col] = min_val
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(bin_img_8, cmap='gray')
plt.show();
```

Ядро

```python
kernel = np.ones((2,2),np.uint8)
```

**Єрозія**

```python
erosion_8 = cv2.erode(bin_img_8, kernel, iterations = 1)
plt.figure(figsize=(5, 5))
plt.imshow(erosion_8, cmap='gray')
plt.show();
```

**Диляція**

```python
dilation_8 = cv2.dilate(bin_img_8, kernel, iterations = 1)
plt.figure(figsize=(5, 5))
plt.imshow(dilation_8, cmap='gray')
plt.show();
```



---


# **2. Методами морфологічної обробки підготувати зображення до подальшої обробки**

```python
img_street = cv2.imread("/content/street.jpg", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(15, 15))
plt.imshow(img_street, cmap='gray')
plt.show();
```

```python
bin_img_street = img_street.copy()
threshold = 170
max_val = 255
min_val = 0
```

```python
for row in range(img_street.shape[0]):
    for col in range(img_street.shape[1]):
        if (img_street[row][col]) > threshold:
            bin_img_street[row][col] = max_val
        else:
            bin_img_street[row][col] = min_val
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(bin_img_street, cmap='gray')
plt.show();
```

```python
erosion_street = cv2.erode(bin_img_street, kernel, iterations = 1)
plt.figure(figsize=(15, 15))
plt.imshow(erosion_street, cmap='gray')
plt.show();
```

**Застосуємо морфологічне відкриття**

```python
opening_street = cv2.morphologyEx(erosion_street , cv2.MORPH_OPEN, kernel)
```

```python
plt.figure(figsize=(25,25))
plt.subplot(121);plt.imshow(opening_street, cmap='gray'); plt.title('Відкриття');
plt.show();
```



---

**Морфологічне закриття**

```python
closing_street = cv2.morphologyEx(erosion_street , cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(25,25))
plt.subplot(121);plt.imshow(closing_street, cmap='gray'); plt.title('Закриття');
plt.show();
```
